```python
import numpy as np
import matplotlib.pyplot as plt
topx=np.array([1, .98, .8, .83, .8, .77, .5, .25, 0, -.25, -.5, -.8, -.8, -.98, -1])

topy=np.array([0, .15, .7, 1.15, 1.03, 1.15, .85, .95, 1, .95, .85, 1.2, .7, .15, 0])
botx=-np.cos(np.pi*np.arange(1,10)/10)
boty=-np.sin(np.pi*np.arange(1,10)/10)
wiskx=np.array([.2, 1.3, .2, 1.4, .2, 1.4, .2, 1.3, .2, .17, .13, .08, .03, 0])
wisky=np.array([ 0, .3, 0, .1, 0, -.1, 0, -.3, 0, .1, -.1, .1, -.1, 0]) -.2
xeye=np.array([0, .2, .3, .4, .43, .45, .43, .4, .37, .35, .37, .4, .5, .6, .5, .4, .3, .2,
0])
yeye=np.array([.5, .5, .43, .4, .42, .5, .58, .6, .58, .5, .42, .4, .43, .5, .57, .6, .57,
.5,.5])-.2
x=np.concatenate((topx,botx,wiskx,xeye,-xeye,-wiskx[::-1]))
y=np.concatenate((topy,boty,wisky,yeye, yeye,wisky[::-1]))
cat = np.array([x,y]).T
plt.plot(cat[:,0], cat[:,1],'b')
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.show()
```

**2) Roatation of matrix**

  a) Anticlockwise

```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.deg2rad(30)   # Convert angle to radians

# Rotation matrix
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])

v = np.array([1, 2])

rotated_v = np.dot(rotation_matrix, v)  # Apply rotation

## plotting
w = rotated_v[0]
s = rotated_v[1]

u = [1]
v = [2]

plt.quiver(0, 0, u, v, color='b', units='xy', scale=1)
plt.quiver(0,0,w,s,color ='r',units ='xy',scale=1)
plt.xlim(-2, 5)
plt.ylim(-2, 2.5)


print("Rotated vector:", rotated_v)
```

```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.deg2rad(-45)  # Convert angle to radians
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])

v = np.array([1, 2])

# Apply rotation
rotated_v = np.dot(rotation_matrix, v)
w = rotated_v[0]
s = rotated_v[1]

## plotting
u = [1]
v = [2]

plt.quiver(0, 0, u, v, color='b', units='xy', scale=1)
plt.quiver(0,0,w,s,color ='r',units ='xy',scale=1)
plt.xlim(0, 5)
plt.ylim(0,5)


print("Rotated vector:", rotated_v)
```
